```python
# Class Methods
# A class method works with the class itself, not a specific object
```

```python
# @classmethod tells Python this is a class method
```

```python
class Student:
    @classmethod
    def show_class(cls):
        print("Class Method")
```

```python
Student.show_class()
```

```python
# Static methods
```

```python
# Static methods are utility methods that do not use self or cls
```

```python
# Works with neither the object nor the clas
```

```python
class Calculator:
    @staticmethod
    def mul(a,b):
        return a*b
```

```python
Calculator.mul(10,20)
```

```python
# Encapsulation
```

```python
# Encapsulation is the process of hiding data using private variables and providing controlled access through methods
```

```python

# Encapsulation protects data by restricting direct access
```

```python
# Use double underscores (__) to make an attribute private
```

```python
class BankAccount:
    def __init__(self,balance):
        self.__balance = balance
    def get_balance(self):
        return self.__balance
```

```python
BankAccount(1000).get_balance()
```

```python
# Single Inheritance
```

```python
# Single inheritance allows one child class to inherit from one parent
```

```python
class Animal:
    def eat(self):
        print("Eating")
class Dog(Animal):
    pass
```

```python
Dog().eat()
```

```python
class Studentmarks:
    def marks(self):
        print("Marks")
class Studentpercent(Studentmarks):
    pass
```

```python
Studentpercent().marks()
```

```python
# Method Overriding
```

```python
# Method overriding replaces a parent method in the child class
```

```python
class Vehicle:
    def start(self):
        print("Vehicle is starting...")
class Car(Vehicle):
    def start(self):
        print("Car is starting...")
```

```python
car = Car()
```

```python
car.start()
```
