```python
# 4.Write a Python program to print numbers from 1 to 50, display all even numbers, and calculate their sum.
```

```python
# Print numbers from 1 to 50
```

```python
sum = 0
for i in range(1, 51):
    if i % 1 == 0:
        print(i)
        sum = sum + i
```

```python
# Display all even numbers
```

```python
sum = 0
for i in range(1, 51):
    if i % 2 == 0:
        print(i)
        sum += i
```

```python
# Calculate their sum
```

```python
print("Sum =", sum)
```

```python
# 5.Write a Python program to generate the multiplication table of a number, display the square of each multiple, and count the total multiples generated.
```

```python
# Multiplication Table
```

```python
num = 10
```

```python
for i in range(1,11):
    print(num,"x",i,"=",num*i)
```

```python
# Display square of each multiple
```

```python
for i in range(1, 11):
    multiple = num * i
    square = multiple ** 2
    print("Multiple =", multiple, "Square =", square)
```

```python
# Count the total multiples generated
```

```python
count = 0
for i in range(1, 11):
    print(num * i)
    count += 1
print("Total multiples generated =", count)
```

```python
# 6.Write a Python program to calculate the factorial of a number, count its digits, and display the reverse of the number.
```

```python
num = int(input("Enter a number: "))
fact = 1
for i in range(1, num + 1):
    fact = fact * i
print("Factorial =", fact)
```

```python
# Count its digits
```

```python
print("Number of digits =", len(num))
```

```python
# Reverse of number
```

```python
print("Reverse =", num[::-1])
```

```python
# 7.Write a Python program to generate the Fibonacci series, calculate its sum, and display the largest term generated.
```

```python
# Generate Fibonacci series
```

```python
n = int(input("Enter number of terms: "))
a = 1
b = 2
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
# Calculate its sum
```

```python
a + b
```

```python
# Display the largest term generated.
```

```python
a = 1
b = 2
for i in range(n):
    largest = a
    a, b = b, a + b
print("Largest term =", largest)
```
