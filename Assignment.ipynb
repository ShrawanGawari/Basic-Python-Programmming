```python
# 11.Write a Python program to reverse a string, count the vowels, and count the consonants.
```

```python
word = "Python"
```

```python
# Reverse a string
```

```python
print(word[::-1])
```

```python
# Count the vowels
```

```python
count = 0
for ch in word.lower():
    if ch in "aeiou":
        count += 1
print("Vowels =", count)
```

```python
# Count the consonents
```

```python
count = 0
for ch in word.lower():
    if ch.isalpha() and ch not in "aeiou":
        count += 1
print("Consonants =", count)
```

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print("Matrix:")
for row in matrix:
    print(row)
```

```python
# 25.Write a Python program to accept N numbers from the user, store them in a list, and display the maximum, minimum, and average values.
```

```python
# Accept N numbers from the user
```

```python
n = int(input("Enter how many numbers: "))
nums = []
for i in range(n):
    num = int(input("Enter number: "))
    nums.append(num)
print("Numbers:", nums)
```

```python
# Store in list
```

```python
print("List:", nums)
```

```python
# Display the maximum, minimum, and average values
```

```python
max(nums)
```

```python
min(nums)
```

```python
avg = sum(nums) / n
```

```python
print(avg)
```
