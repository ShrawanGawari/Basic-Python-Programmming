```python
# break
```

```python
# stop loop at 5
```

```python
for i in range(1,11):
    if i == 5:
        break
    print(i)
    
```

```python
for i in range(1,20):
    if i ==10:
        break
    print(i)
```

```python
for i in range(1,20):
    if i ==10:
        continue
    print(i)
```

```python
while True:
    amount = int(input("enter amount(0 to exit):"))
    if amount == 0:
        break
    print("amount entered:",amount)
    
```

```python
# factorial calculator
```

```python
n = int(input("enter a positive integer:"))
factorial = 1
for i in range(1, n + 1):
    factorial *= i
print(f"factorail of {n} is {factorial}")
```
