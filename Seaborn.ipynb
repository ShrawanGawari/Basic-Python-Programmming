```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("iris")
```

```python
data.head(5)
```

```python
data['species'].nunique()
```

```python
data['species'].unique()
```

```python
sns.kdeplot(data=data, x="sepal_length", fill=True)
```

```python
sns.kdeplot(data=data, x="sepal_length", fill=True)
plt.title("Univariate Distribution")
plt.show()
```

```python
sns.kdeplot(data=data, x="sepal_length", fill=False)
```

```python
sns.kdeplot(
    data= data,
    x="sepal_length",
    y="sepal_width",
    fill= True
)
```

```python
sns.kdeplot(
    data= data,
    x="sepal_length",
    y="sepal_width",
    fill= True
)
plt.title("Bivariate Distribution")
plt.show()
```

```python
sns.jointplot(data=data, x="sepal_length", y="sepal_width", kind="scatter")
```

```python
sns.pairplot(data)
plt.show()
```

```python
data = sns.load_dataset("tips")
```

```python
data
```

```python
sns.pairplot(data)
plt.show()
```

```python
sns.lmplot(
    data=data,
    x="total_bill",
    y="tip"
)
plt.title("Linear Regression")
```

```python
sns.scatterplot( x="total_bill",y="tip",data=data)
```

```python
sns.lineplot( x="total_bill",y="tip",data=data)
```

```python
sns.barplot( x="day",y="total_bill",data=data)
```

```python
sns.countplot( x="day",data=data)
```

```python
sns.boxplot( x="day",y="total_bill",data=data)
```

```python
sns.histplot(data["total_bill"])
```
