```python
# 6.Write a Python program to demonstrate class methods, static methods, and instance methods in the same class.
```

```python
# Class methods,static methods,and instance methods in same class
```

```python
class Student:
    school = "ABC School"   
    def __init__(self, name):
        self.name = name     
    # Instance method
    def show(self):
        print("Name:", self.name)
    # Class method
    @classmethod
    def show_school(cls):
        print("School:", cls.school)
    # Static method
    @staticmethod
    def info():
        print("This is Student class")
# creating object
s1 = Student("Amit")
# instance method (object call)
s1.show()
# class method (class call)
Student.show_school()
# static method (class call)
Student.info()
```

```python
# 7.Write a Python program to implement single inheritance, access parent members, and override a method.
```

```python
# Implement single inheritance, access parent members, and override a method.
```

```python
# Parent class
class Parent:
    def show(self):
        print("This is Parent class")
# Child class (Single Inheritance)
class Child(Parent):
    def show(self):  
        print("This is Child class")
        super().show()   
# object of child class
c1 = Child()
# calling method
c1.show()
```

```python
# 8.Write a Python program to implement multilevel inheritance, access inherited members, and display the inheritance hierarchy.
```

```python
# Implement multilevel inheritance, access inherited members, and display the inheritance hierarchy
```

```python
# Grandparent class
class A:
    def show_A(self):
        print("Class A")
# Parent class
class B(A):
    def show_B(self):
        print("Class B")
# Child class
class C(B):
    def show_C(self):
        print("Class C")
# object of child class
obj = C()
# accessing inherited members
obj.show_A()
obj.show_B()
obj.show_C()
```

```python
# 14.Write a Python program to create an abstract class, implement abstract methods, and instantiate child classes.
```

```python
# Create an abstract class, implement abstract methods, and instantiate child classes.
```

```python
from abc import ABC, abstractmethod
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass
class Dog(Animal):
    def sound(self):
        print("Dog barks")
class Cat(Animal):
    def sound(self):
        print("Cat meows")
d = Dog()
c = Cat()
d.sound()
c.sound()
```

```python
# 15.Write a Python program to demonstrate abstraction using the abc module, implement multiple child classes, and display outputs.
```

```python
# Demonstrate abstraction using the abc module, implement multiple child classes, and display outputs.
```

```python
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
class Rectangle(Shape):
    def area(self):
        print("Rectangle Area: 20")
class Circle(Shape):
    def area(self):
        print("Circle Area: 28.26")
r = Rectangle()
c = Circle()
r.area()
c.area()
```

```python
# 16.Write a Python program to overload arithmetic operators, compare objects, and display the results.
```

```python
# Overload arithmetic operators, compare objects, and display the results.
```

```python
class Number:
    def __init__(self, value):
        self.value = value
    def __add__(self, other):
        return self.value + other.value
    def __gt__(self, other):
        return self.value > other.value
n1 = Number(10)
n2 = Number(5)
print(n1 + n2)
print(n1 > n2)
```

```python
# 23.Write a Python program to design a Student class, store student details, and calculate grades.
```

```python
# Design a Student class, store student details, and calculate grades.
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    def grade(self):
        if self.marks >= 75:
            return "A"
        elif self.marks >= 60:
            return "B"
        elif self.marks >= 40:
            return "C"
        else:
            return "Fail"
s1 = Student("Shrawan", 80)
print(s1.name, s1.marks, s1.grade())
```

```python
# 24.Write a Python program to design an Employee class, calculate salary, and generate employee reports.
```

```python
# Design an Employee class, calculate salary, and generate employee reports.
```

```python
class Employee:
    def __init__(self, name, basic):
        self.name = name
        self.basic = basic
    def salary(self):
        return self.basic + (0.2 * self.basic) + (0.1 * self.basic)
e1 = Employee("Shrawan", 20000)
print(e1.name, e1.salary())
```
