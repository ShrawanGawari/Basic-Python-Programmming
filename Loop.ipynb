```python
for i in range(1,11):
    if i == 5:
        break
    print(i)
```

```python
# ATM exit progran
```

```python
while True:
    amount = int(input("Enter amount(0 to Exit):"))
    if amount == 0:
        break
    print("Amount Entered:",amount)
```

```python
# Factorial calculator
```

```python
n = int(input("Enter a positive integer:"))
factorial = 1
for i in range(1,n+1):
    factorial *= i
print(f"factorial of {n} is {factorial}")
```
