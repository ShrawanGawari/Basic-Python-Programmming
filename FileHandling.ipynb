```python
# File handling in Python is the process of creating, reading, writing, updating, 
# and deleting files using functions like open() 
# and file modes.
```

```python
# Create a Sample File
```

```python
with open("sample.txt", "w") as f:
    f.write("Name: Shrawan\n")
    f.write("Role: Data Scientist\n")
    f.write("Location: India\n")
```

```python
# Reading the File
```

```python
with open("sample.txt", "r") as f:
    data = f.read()
    print(data)
```

```python
# Reading Line by Line
```

```python
with open("sample.txt", "r") as f:
    for line in f:
        print(line.strip())
```

```python
# Appending Data
```

```python
with open("sample.txt", "a") as f:
    f.write("Appended line 1\n")
    f.write("Appended line 2\n")
```

```python
# Read Again After Append
```

```python
with open("sample.txt", "r") as f:
    print(f.read())
```
