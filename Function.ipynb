```python
show(age = 19,name = "Shrawan")
```

```python
# Variable length argument
```

```python
# *args(multiple values)
```

```python
def add(*nums):
    print(sum(nums))
```

```python
add(10,20,30)
```

```python
def multiplication(*arg):
    multi = 1
    for number in arg:
        multi *= number
    return multi
```

```python
multiplication(10,20,30)
```

```python
# **kwargs(key value)
```

```python
def info(**data):
    print(data)
```

```python
info(name="Shrawan",age=19)
```

```python
def add(a,b):
    return a+b
```

```python
result = add(5,3)
```

```python
print(result)
```

```python
add = lambda a,b:a+b
print(add(10,20))
```
